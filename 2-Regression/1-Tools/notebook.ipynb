```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection, __version__
print("sklearn.__version__ is ", __version__)
```

```python
# load diabetes data from the datasets and assign to X,y
diabetes = datasets.load_diabetes()
x = diabetes.data
y = diabetes.target
```

```python
# prepare for plot and regression with numpy new axis
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

```python
#train a linear regression model
# Select just one feature to target for this exercise
X = X[:, np.newaxis, 2]
```

```python
# plot the X_test and y_pred values
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# predictions on test data
y_pred = model.predict(X_test)
```

```python
# display results in the plot  
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.show()
```
